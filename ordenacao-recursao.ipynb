## Problema 1

Crie uma função recursiva para resolver a sequência de Fibonacci..

**Exemplo**

Se a entrada for:


```python
4
```

A saída deve ser:

> 3

````python
def fibonacci(n):

  if n == 1 or n == 0:
    return 1
  else :
    return fibonacci(n-1) + fibonacci(n-2)
````

````python
fibonacci(5)
````

## Problema 2

Crie uma função recursiva para calcular a soma de uma lista de números.

**Exemplo**

Se a entrada for:


```python
[3, 1, 10, 4]
```

A saída deve ser:

> 18

````python
def soma(lista):
  n = len(lista)

  if n == 1:
     return lista[0]
  else:
     return lista[0] + soma(lista[1:])
````

````python
lista = [3, 1, 10]

soma(lista)
````

## Problema 3

Crie uma função recursiva para calcular a soma dos dígitos de um número inteiro positivo.

**Exemplo**

Se a entrada for:


```python
25
```

A saída deve ser:

> 7

**Dica:** Utilize os operadores de divisão `% `e `//`

````python
def soma_digito(n):

  if n == 0:
     return n
  else:
    n1 = n%10
    return n1 + soma_digito(n//10)


````

````python
soma_digito(25)
````

## Problema 4

Crie uma função recursiva para verificar se uma string é um palindromo.

**Exemplo**

Se a entrada for:


```python
'reviver'
```

A saída deve ser:

> True

````python
def palavraPalindromo(p):
  if len(p) <= 1:
    return True
  if p[0] == p[-1]:
    return palavraPalindromo(p[1: 1])
  else:
    return False
````

````python
palavraPalindromo('nada')
````

## Problema 5

Crie uma função recursiva que busca um elemento dentro de uma lista ordenada e retorna o índice desse elemento. Caso o índice não seja encontrado a função deve retornar o valor -1. Essa função precisa estar baseada no algoritmo de busca binária.

**Exemplo**

Se a entrada for:


```python
lista = [1, 5, 7, 12, 35, 49]
alvo = 12
```

A saída deve ser:

> 3

````python
def recursao_binaria(lista, n, comeco=0, final=None): #none significa um valor inesistente ou indefinido, a várialvel n será o elemento

    # ve se o valor de 'final' foi fornecido, se não foi dado joga ele para o fim da lista.
    #verify if the final value exist, else it will be placed at the end of the list
    if final is None:
        final = len(lista) - 1

    # se o começo for maior que o final, n não existe e retorna -1
    # if 'comeco' bigger than 'final', 'n' doesn't exist and it return -1
    if comeco > final:
        return -1

    # calculando o índice do n do meio da lista.
    # calculating the index of the 'n' in the middle list
    meio = (comeco + final) // 2

    # Se o n do meio for igual ao n que estamos procurando, retorne o índice do meio.
    # if the middle element is equal to element we looking, return to the middle index
    if lista[meio] == n:
        return meio

    # Se o n do meio for maior que o n que estamos procurando, faça uma busca recursiva na metade esquerda da lista.
    # If the middle element than element we looking, perform a recursive search in the middle of the list
    elif lista[meio] >= n:
        return recursao_binaria(lista, n, comeco, meio - 1)

    # Se o n do meio for menor que o n que estamos procurando, faça uma busca recursiva na metade direita da lista.
    # if the middle is lees than element we looking, perform a recursive search in the left  of the list
    else:
        return recursao_binaria(lista, n, meio + 1, final)

````

````python
numeros = [1, 5, 7, 12, 35, 49]
alvo = 1
busca = recursao_binaria(numeros, alvo)

print(busca)
````
